```python
from generate_data import *
from baseline_results import *
from attention_model_adam import *
```

```python
# Parameters

num_corefea = 5
corr = 0.1
x_noise = 0.5
shift = True
n_cat = 10
```

```python
### Repeat experiment 20 times, each with different seeds
### Record mse_train, acc_train, mse_test, acc_test of all 5 models (LR, RF, GB, MLP, ATTN)

np.random.seed(2023)
seeds = np.random.randint(10000, size = 20)

results = []

for seed in seeds:
    print('Seed = ', seed)
    train, test = generate_data(N = 2000, D = 5, num_corefea = num_corefea, corr = corr, x_noise = x_noise, 
                                y_noise = 2, n_cat = n_cat, seed = seed, shift = shift)
    attn = train_attention_model(train = train, test = test, n_cat = n_cat, embed_dim = 20, ff_dim = 5, 
                              num_head = 5, seed = seed)
    baseline = get_baseline_results(train = train, test = test, n_cat = n_cat, seed = seed)
    
    baseline['mse_train'].append(attn[0])
    baseline['acc_train'].append(attn[1])
    baseline['mse_test'].append(attn[2])
    baseline['acc_test'].append(attn[3])
    
    results.append(baseline)
```

```python
#### Given results, calculate the mean and std for each metric for all 5 models

mse_train = [x['mse_train'] for x in results]
acc_train = [x['acc_train'] for x in results]
mse_test = [x['mse_test'] for x in results]
acc_test = [x['acc_test'] for x in results]

mse_train_summ = [list(np.array(mse_train).mean(axis=0).round(4)),
                  list(np.array(mse_train).std(axis=0).round(4))]

acc_train_summ = [list(np.array(acc_train).mean(axis=0).round(4)),
                  list(np.array(acc_train).std(axis=0).round(4))]

mse_test_summ = [list(np.array(mse_test).mean(axis=0).round(4)),
                  list(np.array(mse_test).std(axis=0).round(4))]

acc_test_summ = [list(np.array(acc_test).mean(axis=0).round(4)),
                  list(np.array(acc_test).std(axis=0).round(4))]
```

Order of results:
1. Train MSE (LR, RF, GB, MLP, ATN) and SD
2. Train accuracy (LR, RF, GB, MLP, ATN) and SD
3. Test MSE (LR, RF, GB, MLP, ATN) and SD
4. Test accuracy (LR, RF, GB, MLP, ATN) and SD

### Generate result table

```python
result_table = pd.DataFrame()
result_table['seed'] = seeds

cols = ['mse_train', 'acc_train', 'mse_test', 'acc_test']
models = {'lr': 0, 'rf': 1, 'gb': 2, 'mlp': 3, 'at': 4}

for col in cols:
    for model in models:
        result_table[col + '_' + model] = [x[col][models[model]] for x in results]
        
table_name = str(num_corefea) + '_' + str(corr) + '_' + str(x_noise) + '_' + str(shift) + '_' + str(n_cat)
result_table.to_csv(table_name + '.csv')
```
